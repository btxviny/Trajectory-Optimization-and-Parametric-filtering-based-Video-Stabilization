```python
import subprocess
import os 
import pandas as pd
from metrics import metric
import warnings
warnings.filterwarnings("ignore")
```

```python
scripts = [os.path.join('./scripts/',x) for x in os.listdir('scripts')]
scripts = scripts[1:]
scripts
```

```python
scripts[0].split('/')[-1].split('.')[0]
```

```python
root = './data'
results = './results'
categories = os.listdir(root)
scores_dict = {
    'Category': [],
    'Cropping Score': [],
    'Distortion Score': [],
    'Stability Score': [],
    'Pixel Score': []
}
```

```python
for SCRIPT_PATH in scripts:
    for category in categories:
        cropping_score = 0
        distortion_score = 0
        stability_score = 0
        pixel_score = 0
        category_name = category
        category = os.listdir(os.path.join(root,category))
        for video in category:
            in_path = os.path.join(root,category_name,video)
            out_path = os.path.join(results,f'{category_name}_{video}')
            subprocess.run(['python', SCRIPT_PATH, '--in_path', in_path, '--out_path', out_path])
            cropping,distortion,stability, pixel = metric(in_path,out_path)
            cropping_score += cropping
            distortion_score += distortion
            pixel_score += pixel
            stability_score += stability
        cropping_score /= len(category)
        distortion_score /= len(category)
        pixel_score /= len(category)
        stability_score /= len(category)
        scores_dict['Category'].append(category_name)
        scores_dict['Cropping Score'].append(cropping_score)
        scores_dict['Distortion Score'].append(distortion_score)
        scores_dict['Stability Score'].append(stability_score)
        scores_dict['Pixel Score'].append(pixel_score)
    df = pd.DataFrame(scores_dict)
    file_name = SCRIPT_PATH.split('/')[-1].split('.')[0] + '.csv'
    df.to_csv(file_name, index=False)
```
